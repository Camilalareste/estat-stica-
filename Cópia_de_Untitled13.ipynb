## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Create new categorical columns based on cost and time intervals, and convert categorical columns to binary representations. Then, handle any missing values in these new columns.



```python
# Categorize Cost
def categorize_cost(cost):
    if 0 <= cost <= 5:
        return '0-5'
    elif 5 < cost <= 10:
        return '5-10'
    elif 10 < cost <= 20:
        return '10-20'
    elif cost > 20:
        return '20+'
    else:
        return None  # Handle NaN or other invalid values

df['Custo_Intervalo'] = df['Custo_Diario_Transporte'].apply(categorize_cost)

# Categorize Time
def categorize_time(time):
    if time < 1:
        return '<1 hour'
    elif 1 <= time <= 2:
        return '1-2 hours'
    elif time > 2:
        return '>2 hours'
    else:
        return None  # Handle NaN or other invalid values

df['Tempo_Intervalo'] = df['Tempo_Deslocamento'].apply(categorize_time)

# Binary Encoding
def binary_encode(value):
    if isinstance(value, str):
        if 'sim' in value.lower() or 'Sim' in value:
            return 1
        elif 'não' in value.lower() or 'Não' in value:
            return 0
        else:
            return None
    else:
        return None

df['Custo_Influencia_Binario'] = df['Custo_Influencia_Frequencia'].apply(binary_encode)
df['Beneficio_Binario'] = df['Beneficio_Estudantil_Transporte'].apply(binary_encode)


# Handle Missing Values (using mode imputation)
for col in ['Custo_Intervalo', 'Tempo_Intervalo', 'Custo_Influencia_Binario', 'Beneficio_Binario']:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)

display(df.head())
display(df.info())
```

## Data analysis

### Subtask:
Statistical Analysis of Transportation Data


**Reasoning**:
Perform the statistical analysis as requested in the subtask, including hypothesis testing, correlation analysis, and t-tests or ANOVA.



```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.proportion import proportions_ztest
import scipy.stats as stats
import numpy as np

# Hypothesis Testing (One-sample proportion z-test)
# Null hypothesis: Proportion of students using public transport = 0.5
# Alternative hypothesis: Proportion of students using public transport != 0.5

public_transport_count = len(df[df['Meio_Transporte'].str.contains('ônibus|transporte público', case=False, na=False)])
total_students = len(df)
count = np.array([public_transport_count])
nobs = np.array([total_students])
z_stat, p_value = proportions_ztest(count, nobs, value=0.5, alternative='two-sided')

print("One-sample proportion z-test:")
print(f"Test Statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.3f}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis.  The proportion of students using public transport is significantly different from 50%.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in the proportion of students using public transport from 50%.")

# Correlation Analysis
correlation_coefficient, p_value = stats.pearsonr(df['Custo_Diario_Transporte'], df['Custo_Influencia_Binario'])
print("\nCorrelation Analysis:")
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.3f}")
print(f"P-value: {p_value:.3f}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a statistically significant correlation between transportation cost and attendance frequency.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no statistically significant correlation between transportation cost and attendance frequency.")


# t-tests or ANOVA (Comparing average transportation cost between transport methods)

# Perform ANOVA
model = ols('Custo_Diario_Transporte ~ C(Meio_Transporte)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nANOVA:")
print(anova_table)

# Perform post-hoc test (Tukey's HSD)
tukey_result = pairwise_tukeyhsd(df['Custo_Diario_Transporte'], df['Meio_Transporte'], alpha=0.05)
print("\nTukey's HSD Post-Hoc Test:")
print(tukey_result)

# Limitations of Small Sample Size
print("\nLimitations of Small Sample Size:")
print("The small sample size (n=15) limits the generalizability of these findings to the broader student population.  The results may not accurately reflect the transportation patterns and costs of all economics students at UFRPE.")
print("A larger sample size would increase the statistical power of the tests, leading to more reliable conclusions and a reduction in the risk of Type II errors (failing to detect a true effect).")
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.proportion import proportions_ztest
import scipy.stats as stats
import numpy as np

# Hypothesis Testing (One-sample proportion z-test)
# Null hypothesis: Proportion of students using public transport = 0.5
# Alternative hypothesis: Proportion of students using public transport != 0.5

public_transport_count = len(df[df['Meio_Transporte'].str.contains('ônibus|transporte público', case=False, na=False)])
total_students = len(df)
count = np.array([public_transport_count])
nobs = np.array([total_students])
z_stat, p_value = proportions_ztest(count, nobs, value=0.5, alternative='two-sided')

print("One-sample proportion z-test:")
print(f"Test Statistic: {float(z_stat):.3f}") # Convert to float before formatting
print(f"P-value: {float(p_value):.3f}") # Convert to float before formatting
if float(p_value) < 0.05:
    print("Conclusion: Reject the null hypothesis.  The proportion of students using public transport is significantly different from 50%.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in the proportion of students using public transport from 50%.")

# Correlation Analysis
correlation_coefficient, p_value = stats.pearsonr(df['Custo_Diario_Transporte'], df['Custo_Influencia_Binario'])
print("\nCorrelation Analysis:")
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.3f}")
print(f"P-value: {p_value:.3f}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a statistically significant correlation between transportation cost and attendance frequency.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no statistically significant correlation between transportation cost and attendance frequency.")


# t-tests or ANOVA (Comparing average transportation cost between transport methods)

# Perform ANOVA
model = ols('Custo_Diario_Transporte ~ C(Meio_Transporte)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nANOVA:")
print(anova_table)

# Perform post-hoc test (Tukey's HSD)
tukey_result = pairwise_tukeyhsd(df['Custo_Diario_Transporte'], df['Meio_Transporte'], alpha=0.05)
print("\nTukey's HSD Post-Hoc Test:")
print(tukey_result)

# Limitations of Small Sample Size
print("\nLimitations of Small Sample Size:")
print("The small sample size (n=15) limits the generalizability of these findings to the broader student population.  The results may not accurately reflect the transportation patterns and costs of all economics students at UFRPE.")
print("A larger sample size would increase the statistical power of the tests, leading to more reliable conclusions and a reduction in the risk of Type II errors (failing to detect a true effect).")
```
